Methode de regression lineaire 

```python
from sklearn.linear_model import LinearRegression

numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

for i in range(len(numeric_columns)):
    for j in range(i + 1, len(numeric_columns)):
        feature1 = numeric_columns[i]
        feature2 = numeric_columns[j]

        # Créer un sous-ensemble du DataFrame avec les deux caractéristiques
        subset_df = data[[feature1, feature2]].dropna()

        # Ajuster le modèle de régression linéaire
        model = LinearRegression()
        model.fit(subset_df[[feature1]], subset_df[feature2])

    
    
    # Récupérer le coefficient de détermination (R^2)
        r_squared = model.score(subset_df[[feature1]], subset_df[feature2])

        # Afficher les résultats si R^2 >= 0.9
        if r_squared >= 0.9:
            print(f"Régression linéaire entre {feature1} et {feature2} - R^2 : {r_squared:.4f}")

            # Tracé de la régression linéaire pour visualisation
            plt.scatter(subset_df[feature1], subset_df[feature2], label='Données')
            plt.plot(subset_df[feature1], model.predict(subset_df[[feature1]]), color='red', label='Régression linéaire')
            plt.xlabel(feature1)
            plt.ylabel(feature2)
            plt.legend()
            plt.show()
```

### OUTLIERS

```python
# Select numerical columns
numerical_data = data.select_dtypes(include=['int64', 'float64'])

# Function to detect outliers using IQR method
def detect_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers

# Detect outliers for each numerical column
outliers_dict = {}
for column in numerical_data.columns:
    outliers = detect_outliers_iqr(numerical_data[column])
    outliers_dict[column] = outliers
    if outliers.sum() > 0:
        # Create a boxplot for columns with outliers
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=data[column])
        plt.title(f'Boxplot for {column} (Outliers: {outliers.sum()})')
        plt.xlabel(column)
        plt.show()
```

```python
seuil_outliers = 1.5

for column in ['Count', 'srv_count']:
    if np.issubdtype(data[column].dtype, np.number):
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        outliers_indices = ((data[column] < (Q1 - seuil_outliers * IQR)) | (data[column] > (Q3 + seuil_outliers * IQR)))

        # Imputation des valeurs aberrantes par la moyenne
        mean_value = data[~outliers_indices][column].mean()
        data.loc[outliers_indices, column] = mean_value

        # Calcul du nombre de valeurs aberrantes après imputation
        new_outliers_indices = ((data[column] < (Q1 - seuil_outliers * IQR)) | (data[column] > (Q3 + seuil_outliers * IQR)))
        num_new_outliers = new_outliers_indices.sum()

        print(f"Colonne {column} - Imputation des valeurs aberrantes par la moyenne")
        print(f"Colonne {column} - Nombre de valeurs aberrantes avant imputation :", outliers_indices.sum())
        print(f"Colonne {column} - Nombre de valeurs aberrantes après imputation :", num_new_outliers)

# Vous pouvez également sauvegarder le DataFrame modifié si nécessaire
# data.to_csv('donnees_apres_imputation_moyenne.csv', index=False)
```

```python
def assignClass(atkType: str) -> int:

    Dos = ["neptune", "back", "land", "pod", "smurf", "teardrop", "mailbomb", "apache2", "processtable", "udpstorm", "worm"]
    U2R = ["buffer_overflow", "loadmodule", "perl", "rootkit", "ps", "sqlattack", "xterm"]
    R2L = ["ftp_write", "guess_passwd", "imap", "multihop", "phf", "spy", "warezclient", "warezmaster",
    "sendmail", "named", "snmpgetattack", "snmpguess", "xlock", "xsnoop", "httptunnel"]
    Probe = ["ipsweep", "nmap", "portsweep", "satan", "mscan", "saint"]


    attack_type = 0  # normal

    if atkType in Dos:
        attack_type = 1
    elif atkType in U2R:
        attack_type = 2
    elif atkType in R2L:
        attack_type = 3
    elif atkType in Probe:
        attack_type = 4

    return attack_type
```

```python
def addGroup(data):
    data['xAttack_g'] = data['xAttack'].apply(assignClass)

on_train_and_test(addGroup)
```

* Evaluation methods (Function) 

```python
from sklearn.metrics import accuracy_score, confusion_matrix,precision_recall_fscore_support, classification_report, roc_curve, auc



def evaluate_model(model, X_train, y_train, X_test, y_test, class_labels):
    # 3. Training the model
    model.fit(X_train, y_train)

    #4. predictions on the test set
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    # Evaluating the model on the test set
    acc_train = accuracy_score(y_train, y_pred_train)
    acc_test = accuracy_score(y_test, y_pred_test)

    conf_matrix = confusion_matrix(y_test, y_pred_test)
    report = classification_report(y_test, y_pred_test)

    # Calculate precision, recall, and F1-score for each class
    precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred_test, average='weighted')

    print("Model Evaluation:")
    print("Accuracy on Train Set:", acc_train)
    print("Accuracy on Test Set:", acc_test)
    print("Precision (weighted):", precision)
    print("Recall (weighted):", recall)
    print("F1-Score (weighted):", f1_score)
    print("Classification Report:\n", report)

    # Plotting the confusion matrix
    plot_confusion_matrix(conf_matrix, class_labels)

def plot_confusion_matrix(conf_matrix, class_labels):
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
    plt.xlabel('Predictions')
    plt.ylabel('True Labels')
    plt.title('Confusion Matrix')
    plt.show()

class_labels = ['Normal', 'Dos', 'U2R', 'R2L', 'Probe']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report

# Assuming you have the classification reports for each model
report_knn = classification_report(y_test, y_predK)
report_svm = classification_report(y_test, y_predsvm)
report_lr = classification_report(y_test, y_predlr)
report_rf = classification_report(y_test, y_predrf)
report_ad = classification_report(y_test, y_predad)

# Extracting metrics for each attack type and each model
def extract_metrics(report):
    lines = report.split('\n')[2:-5]
    precisions = [float(line.split()[1]) for line in lines]
    recalls = [float(line.split()[2]) for line in lines]
    f1_scores = [float(line.split()[3]) for line in lines]
    return precisions, recalls, f1_scores

# Extract metrics values for each model
precision_knn, recall_knn, f1_knn = extract_metrics(report_knn)
precision_svm, recall_svm, f1_svm = extract_metrics(report_svm)
precision_lr, recall_lr, f1_lr = extract_metrics(report_lr)
precision_rf, recall_rf, f1_rf = extract_metrics(report_rf)
precision_ad, recall_ad, f1_ad = extract_metrics(report_ad)

# Attack types
attack_types = ['Normal', 'DoS', 'U2R', 'R2L', 'Probe']

# Model names
model_names = ['KNN', 'SVM-RBF', 'Logistic Regression', 'Random Forest', 'AdaBoost']

# Create a grouped bar chart for each metric
def plot_metrics(metric_values, metric_name, model_name):
    fig, ax = plt.subplots(figsize=(10, 6))

    bar_width = 0.15
    bar_positions = range(len(attack_types))

    for i, metric in enumerate(metric_values):
        ax.bar([p + bar_width*i for p in bar_positions], metric, width=bar_width, label=f'{model_name[i]} - {metric_name} {i+1}')

    ax.set_xticks([p + bar_width for p in bar_positions])
    ax.set_xticklabels(attack_types)
    ax.set_xlabel('Attack Types')
    ax.set_ylabel(metric_name)
    ax.set_title(f'{metric_name} for Each Model and Attack Type')
    ax.legend()

    plt.show()

# Plot for precision
plot_metrics([precision_knn, precision_svm, precision_lr, precision_rf, precision_ad], 'Precision', model_names)

# Plot for recall
plot_metrics([recall_knn, recall_svm, recall_lr, recall_rf, recall_ad], 'Recall', model_names)

# Plot for F1-score
plot_metrics([f1_knn, f1_svm, f1_lr, f1_rf, f1_ad], 'F1-Score', model_names)
```
